```python
import pandas as pd
```

```python
data = [100.67, 102, 104]

series= pd.Series(data , index= ['a', 'b', 'c'])
print(series)
print(series.loc['a']) #label based indexing
print(series.iloc[1]) #integer position
print(series['a']) #shortcut for label based indexing
series['b']= 105
print(series)
```

```python
s = pd.Series([10, 20, 30], index=[1, 2, 3])

s[1]      # could be ambiguous in some contexts
s.loc[1]  # explicitly label 1 → 10 ✅
s.iloc[1] # explicitly position 1 → 20 ✅
```

```python
calories= { "Day 1": 1420, "Day 2": 2380, "Day 3": 1390 , "Day 4": 1720, "Day 5": 1900 }
#no need to pass index as it will be taken from the keys of the dictionary
series= pd.Series(calories)
print(series)


series.loc["Day 1"] += 200
print(series.loc["Day 1"])

print(series[series > 1800]) #values greater than 1500
```

```python
data= {
    "Name" : ["Spongebob", "Patrick", "Squidward"],
    "Age" : [30, 31, 49]
}

df= pd.DataFrame(data, index= ["Employee 1", "Employee 2", "Employee 3"])
print(df)

print(df.loc["Employee 1"]) #accessing row by label
print(df.iloc[1]) #accessing row by integer position
```

```python
df["Job"] = ["Fry Cook", "Starfish", "Cashier"]
print(df)
#new column
```

```python
#new row 

new_row= pd.DataFrame([{"Name": "Sandy", "Age": 32, "Job": "Scientist"}], index= ["Employee 4"] )
df= pd.concat([df, new_row])
print(df)

```

```python
#new rows 

new_rows= pd.DataFrame([
    {"Name": "Chow", "Age": 33, "Job": "Quantitative Analyst"},
    {"Name": "Mr. Krabs", "Age": 70, "Job": "Business Owner"},
], index= ["Employee 5", "Employee 6"] )

df= pd.concat([df, new_rows])
print(df)
```

```python
df1= pd.read_csv("/Users/atharvdeshpande/Desktop/coding /pokemon.csv", index_col= "Name")
print(df1.head())

df1= df1.drop(columns= ["Generation", "Legendary", "#"])
```

```python
print(df1.head())
```

```python
#NaN of the type 2 is replaced by the string "None"

df1= df1.fillna({"Type 2": "None"})
print(df1.iloc[0:10].to_string())
```

```python
#replacing 

df1["Type 1"]= df1["Type 1"].replace({"Water": "H2O"})
print(df1[df1["Type 1"] == "H2O"].iloc[0:10].to_string())
```

```python
#fix data types 

df1["Speed"]= df1["Speed"].astype(float)
print(df1.dtypes)
```
